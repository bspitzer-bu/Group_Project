```python
FILELIST = 'C:/Users/Brian/Desktop/Boston University/Databases/Group_Project/FASTAs/distincts.csv'

import pandas
import csv

filenames = list(pandas.read_csv(FILELIST)['PDB ID'])
genenames = list(pandas.read_csv(FILELIST)['gene name'])
```

```python
OUTFILE = 'C:/Users/Brian/Desktop/Boston University/Databases/Group_Project/FASTAs/fastas2.txt'

for i in range(len(filenames)):
    FASTAFILE = 'C:/Users/Brian/Desktop/Boston University/Databases/Group_Project/FASTAs/Group_Project-master/Group_Project-master/src/media/fastas/'+filenames[i]+'.fasta'
    with open(FASTAFILE) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        seq = next(csv_reader)
        
    csv_file.close()
    
    with open(OUTFILE, 'a') as writeFile:
        writer = csv.writer(writeFile)
        hdr = '>'+genenames[i]
        writer.writerows([[hdr]])
        writer.writerows([seq])
        writer.writerows('')

    writeFile.close()

```
